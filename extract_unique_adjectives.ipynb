```python
from collections import Counter
from konlpy.tag import Okt
import os
import numpy as np
# @@@연예인의 크롤링한 기사제목 파일 --> 읽기전용으로 호출함
files = []
text_files = os.listdir('./data/')

print(text_files)
```

```python
if '.DS_Store' in text_files:
    text_files.remove('.DS_Store')
print(text_files)
```

```python
# 2. 배열 news에 전체댓글을 다시저장(한줄씩)
def create_adj_list(text_file):
    f = open('./data/' + text_file, 'r', encoding='utf-8')
    
    lines = f.readlines()
    news = []

    for line in lines:
        news.append(line)
    
    print(text_file,'>>',len(news))
    f.close()
    
    # 3. 트윗터 패키지 안에 konlpy 모듈호출
    twitter = Okt()

    # 각 문장별로 형태소 구분하기
    sentences_tag = []
    for sentence in news:
        # morpth - 형태소별로 나눠져서 저장됨.
        morph = twitter.pos(sentence)
        sentences_tag.append(morph)
        #print('morph:', len(morph))
        #print(morph)
        #print('-'*30,'다음기사제목입니당')

    #print(sentences_tag)
    print(text_file+'>>', len(sentences_tag))
    #print('\n'*3)

    # 형용사인 품사만 선별해 리스트에 담기 ( 명사도 포함해도 괜찮을듯? 아닌가?)
    adj_list = []
    noun_list = []
    for sentence1 in sentences_tag:
        for word, tag in sentence1:
            # 형용사인것만 adj_list에 담겠다.
            if tag in ['Adjective']:
                adj_list.append(word)
            if tag in ['Noun']:
                noun_list.append(word)


    # 빈도수 계산 & 상위 빈도 10위 까지 출력
    counts = Counter(adj_list)
    return adj_list, noun_list
    #print(counts.most_common(10))
```

```python
unique_adj = []
unique_noun = []

for i in range(0, len(text_files)):
    adj_list, noun_list = create_adj_list(text_files[i])
    
    adj_counter = Counter(adj_list)
    adjectives = adj_counter.keys()
    
    for adj in adjectives:
        unique_adj.append(adj)
        
    
    noun_counter = Counter(noun_list)
    nouns = noun_counter.keys()
    for n in nouns:
        unique_noun.append(n)
```

```python
unique_adj = list(set(unique_adj))
unique_noun = list(set(unique_noun))
```
